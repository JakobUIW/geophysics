# Interactive Forward Modeller
This notebook shows how to use the `InteractiveForwardTEM` class.
### Imports
Start with the following import:

```python
from gp_tools.tem import InteractiveTEM
```

### Initialisation
After importing the `InteractiveForwardTEM`-class, you need to create an instance of it by assigning the class a name. A sample name is seen in the cell below. In order to start using the Interface, the method `.start()` is called. Now the User-Interface should appear below.

### Using the UI
In the UI, you can select the thickness and resistivity of each layer, add layers and substract them. Additionally you can save for each state the image, the depth and resistivities model or the response (as a tem file for interoperability with existing inversion algorithms). You can also save all of them. For your convenience and a better overview of different states, you can give each file a name of your choice. However, a suggestion, based on your name and the project name is provided. Two subsurface models are provided for you as buttons, but you can also load your previously saved ones.

### Using the Backend
As the `InteractiveForwardTEM`-class is a daughter class of the `ForwardTEM`, all properties can be used within a python cell. Just enter your commands as adressed in the `manual_forward_solver.ipynb` before you use the `.start()`-Method.
If necessary, you could make changes after the start of the UI. This might result in bugs, as not all Inputs are filtered. In order to then bring the user interface and model up to date to your changes, you can simply click the `Refresh`-Button or add a command `InteractiveForwardTEM.update()`.

### Error: Too many images just loaded
Should you load a model and more than one image loads, please click the `Refresh`-Button. This is a bug from a module we use, an could not be fixed within the class.

### More Information
Should you be curious, you can look through the `interactive.py` and `modeller.py` Files. They are not fully documented but written rather legibly, so you might gain insight.
