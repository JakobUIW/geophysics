# Interactive Forward Modeller
This notebook shows how to use the `InteractiveForwardTEM` class.
### Imports
Start with the following import:

```python
from gp_tools.tem.forward import JupyterInteractiveTEM
```

### Initialisation
After importing the `InteractiveTEMUpdater`-class, you need to create an instance of it by assigning the class a name. A sample name is seen in the cell below. In order to start using the Interface, the method `.start()` is called. Now the User-Interface should appear below. Please just fill in the `.author` and the `.project_name` -properties.

### Using the UI
In the UI, you can select the thickness and resistivity of each layer. Additionally you can save for each state the image, the depth and resistivities model or the response (as a .tem file for interoperability with existing inversion algorithms). You can also save all of them at once. The files are saved with your author-name, the project-name and a time-stamp to create unique filenames. Two subsurface models are provided for you as buttons.
Furthermore, you can change settings from the instrument (TEM-FAST 48), to achieve different resolutions, investigation depths, and more.

### Updating the model
In order to view the sounding for your current electrical subsurface model, just execute the cell with the `.update()`. This is also the place where you can change the layer number - just type in your prefered number. Executing the `.update()` renews the user interface, but more importantly the plot. (This is necessary because of Jupyter pecularities.)

### Using the Backend
As the `InteractiveTEMUpdater`-class is a daughter class of the `ForwardTEM`, all properties can be used within a python cell. Just enter your commands as adressed in the `manual_forward_solver.ipynb` before you use the `.start()`-Method.
If necessary, you could make changes after the start of the UI. This might result in bugs, as not all Inputs are filtered. In order to then bring the user interface and model up to date to your changes, you can simply click the `Refresh`-Button or add a command `InteractiveTEMUpdater.update()`.


### More Information
